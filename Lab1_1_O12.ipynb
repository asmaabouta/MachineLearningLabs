### Afficher une photo aléatoire a partir d'une liste

```python
from random import choice
from IPython.display import Image,display
list = ["Barack_Obama","Donald_Trump","Ivanka_Trump","Michelle_Obama"]
name = choice(list)
print ("Name : ",name)
display(Image(filename="photos-O12/"+name+".jpg"))
```

```python
from random import choice
from IPython.display import Image,display
list = ["Barack_Obama","Donald_Trump","Ivanka_Trump","Michelle_Obama"]
x=''
while list and x !='q':
    name = choice(list)
    print ("Name : ",name)
    display(Image(filename="photos-O12/"+name+".jpg"))
    list.remove(name)
    x = input("Enter to continue or q to Exit")
print("-------END---------")
```

### choisir un nom aléatoire à partir d'un dossier

```python
from random import choice
from IPython.display import Image,display
import glob
list = glob.glob("photos-O12/*.jpg")
#list
x=''
while list and x !='q':
    name = choice(list)
    print ("Name of photo: ",name)
    display(Image(filename = name))
    list.remove(name)
    x = input("Enter to continue or q to Exit")
print("-------END---------")
```
